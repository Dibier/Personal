**1)** Diga qué índices sería pertinente tener y póngalos.

```python
#Para mí los índices pertinentes corresponden al número a  year, month, day, hour

dat = pd.read_csv('~/Jupyter Notebooks/PRSA_data_2010.1.1-2014.12.31.csv', index_col=[1, 2, 3, 4])
display(dat)

```

**2)**  Elimine registros que tengan NaN en la columna "pm2.5"

```python
#Creamos un filtro de filas NaN en la columna pm2.5"

filtro = dat['pm2.5'].notna() 

dat = dat[filtro]

display(dat)
```

**3)** Realice un describe de todas las variables por año. Además de realizar un gráfico como boxplot con el que pueda comparar los resultados dados.

```python
dat0.groupby('year')[dat0.columns].describe()

```

```python
for i in dat.columns:
    if i != 'cbwd':
        dat.boxplot(by='year', column=[i], grid=False)
        plt.show()

```

**4)** Realice un gráfico de disperción de las variables "pm2.5" y "TEMP" con título, labels apropiados y color rojo para los datos. Describa: ¿Qué se puede inferir de dicho gráfico?

```python

x = np.array(dat['pm2.5'])
y = np.array(dat['TEMP'])

fig = plt.figure() #Crea una figura

axes = fig.add_axes([4.1,0.1, 0.8, 0.8])

axes.scatter(x, y, color ='r', marker='o', alpha = 0.2) #Gráfico de dispersión con transparencia
axes.set_xlabel('pm2.5')
axes.set_ylabel('TEMP')
axes.set_title('Gráfico de dispersión pm2.5 vs TEMP')

plt.show()
```

***Conclusión:*** En el gráfico, los puntos se aglomeran en una región que comprende el eje 'TEMP' y está ubicada por debajo antes de pm2.5 = 600.
